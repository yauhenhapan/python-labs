```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("E:\Study\Python\python-labs\Lab4\la-crimes-sample.csv", delimiter=",")
data_human_origins = pd.read_csv("E:\Study\Python\python-labs\Lab4\human_origins.csv", delimiter=",")
```

```python
# Count of rows and columns
rows = len(data)
cols = len(data.columns)
print("Number of Rows: ", rows)
print("Number of Columns: ", cols)
```

```python
# Names of columns
columns_names = data.columns.tolist()
columns_names
```

```python
# Data type of columns
dataTypes = data.dtypes 
dataTypes
```

```python
# Count of unique values in tables
for col in data.columns:
    print(("column: " + col + ", count={0}, unique={1}").format(data[col].count(), data[col].nunique()))
```

```python
# Count of missing values in columns
total = 0
for col in data.columns:
    count_missing_values = data[col].isnull().sum()
    total += count_missing_values
    print(col, ":", count_missing_values)
print("total =", total)
```

```python
# Distribution of numeric variables
data.describe()
```

```python
# Is it true that women are more often victims than men?
count_men = 0
count_women = 0
for val in data["Victim Sex"]:
    if val == "M":
        count_men += 1
    if val == "F":
        count_women += 1
print("Count men = ", count_men)
print("Count women = ", count_women)
if(count_women > count_men):
    print("True")
else:
    print("False")
```

```python
# Study the distribution of the number of crimes by age. What's the trend? What age people are most often subjected to crimes? Are there local minimums?
victims_age = data["Victim Age"].value_counts().reset_index()
# Take top 20
va = victims_age.head(20)
print("People most often subjected to crimes:", victims_age.head(1))
plt.hlines(y=va.index, xmin=0, xmax=va['Victim Age'],  linewidth=3)
plt.gca().set(ylabel='Victim Age', xlabel='Count of victims')
plt.yticks(va.index, va['index'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
arr = data['Victim Age'].value_counts().reset_index()
print(arr)
plt.plot(arr['index'],arr['Victim Age'])
plt.gca().set(ylabel='Number of crimes', xlabel='Victim Age');
```

```python
# How does the likelihood of women and men to become victims differ by age? Build the visualization. In what age range are men more likely to become victims of crime?
data[["Victim Age", "Victim Sex"]].pivot_table(index=["Victim Age"], columns=["Victim Sex"], aggfunc=len).plot();
```

```python
data[["Victim Age", "Victim Descent"]].pivot_table(index=["Victim Age"], columns=["Victim Descent"], aggfunc=len).plot();
```

```python
# The 10 most common crimes in Los Angeles. Build a graph.
most_common_crimes_districts = data['Crime Code Description'].value_counts().reset_index().head(10)
plt.hlines(y=most_common_crimes_districts.index, xmin=0, xmax=most_common_crimes_districts['Crime Code Description'],  linewidth=3)
plt.gca().set(ylabel='Crimes Description')
plt.yticks(most_common_crimes_districts.index, most_common_crimes_districts['index'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
# From what crimes do women most often try, and from what kind of men?
# Women, top 20 crimes
arr_women = data[data['Victim Sex'].isin(['F'])];
data_women = arr_women['Crime Code Description'].value_counts().reset_index().head(20)
plt.hlines(y=data_women.index, xmin=0, xmax=data_women['Crime Code Description'],  linewidth=5)
plt.gca().set(ylabel='Crimes Description')
plt.yticks(data_women.index, data_women['index'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
# From what crimes do women most often try, and from what kind of men?
# Men, top 20 crimes
arr_men = data[data['Victim Sex'].isin(['M'])];
data_men = arr_men['Crime Code Description'].value_counts().reset_index().head(20)
plt.hlines(y=data_men.index, xmin=0, xmax=data_men['Crime Code Description'],  linewidth=5)
plt.gca().set(ylabel='Crimes Description')
plt.yticks(data_men.index, data_men['index'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
# What kind of people are most often subjected to crimes?
# top 10
data_result = data['Victim Descent'].value_counts().reset_index().head(10)
data_result = data_result.merge(data_human_origins)
print(data_result)
plt.hlines(y=data_result.index, xmin=0, xmax=data_result['Victim Descent'],  linewidth=3)
plt.gca().set(ylabel='Victim Descent')
plt.yticks(data_result.index, data_result['Descent Description'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
# Sort the districts by the number of crimes. Build a graph showing the safest and most dangerous areas.
districts = data['Area Name'].value_counts().reset_index()
print(districts)
plt.hlines(y=districts.index, xmin=0, xmax=districts['Area Name'],  linewidth=5)
plt.gca().set(ylabel='$Area Name$')
plt.yticks(districts.index, districts['index'], fontsize=10)
plt.grid(linestyle='--', alpha=0.5)
```

```python
# People of what origin most often suffer from crime in each of the areas
arr_people = data['Area Name'].value_counts().reset_index()
res_people = pd.DataFrame(columns=['index', 'Victim Descent', 'Area Name'],index=range(0))
for index, row in arr_people.iterrows():
    an = data[data['Area Name'] == row['index']]
    res = an['Victim Descent'].value_counts().reset_index().head(1)
    res_people = res_people.append(res)
    res_people['Area Name'].iloc._setitem_with_indexer(index, row['index'])
res_people = res_people.merge(data_human_origins)
res_people = res_people.drop('index', 1)
res_people = res_people.drop('Victim Descent', 1)
print(res_people)
```
